Figure S5 in Mahdavi et al. (2021) Environmental Pollution (in the Supplemental Information (SI))

These figures illustrate times-eries DustTrak PM2.5 and TSP concnetrations in the course of a week in Weeks 2, 3, 4, 5 and 6 (respectively) in the studied home. Week 1 data is not shown as the DustTrak didn't perform for the entire week.

Plot type: Box plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import timedelta
import datetime as dt
exec(open(r'C:\PhD Research\Generic Codes\notion_corrections.py').read())
```

```python
df = pd.read_excel(backslash_correct(r'C:\PhD Research\Airborne\Processed\\') + 'dt_drx.xlsx')

df = df[['Date', 'PM2.5', 'TSP', 'visit']]

for v in [2,3,4,5,6]:
    df_s = df[df['visit'] == v]
    min_ = df_s['Date'].min()
        
    plt.figure()
    plt.plot(df_s['Date'], df_s['PM2.5'], c = 'b', lw = 0.5, label = 'PM$_{2.5}$')
    plt.plot(df_s['Date'], df_s['TSP'], c = 'r', lw = 0.5, label = 'TSP')
    
    plt.xlim(df[df['visit'] == v]['Date'].min(), df[df['visit'] == v]['Date'].max())
    plt.xticks([])
        
    plt.yscale('log')
    plt.ylim(1,1000)
    plt.yticks(np.logspace(0,3,4))
    plt.ylabel('Concentration (µg/m$^{3}$)', size = 14)
    # plt.tick_params(axis = 'y', direction = 'inout')
    
    plt.axvline(x = min_ + timedelta(minutes = 1440), c = 'k', ls = (0, (20, 6)), lw = 0.5)
    plt.axvline(x = min_ + timedelta(minutes = 2*1440), c = 'k', ls = (0, (20, 6)), lw = 0.5)
    plt.axvline(x = min_ + timedelta(minutes = 3*1440), c = 'k', ls = (0, (20, 6)), lw = 0.5)
    plt.axvline(x = min_ + timedelta(minutes = 4*1440), c = 'k', ls = (0, (20, 6)), lw = 0.5)
    plt.axvline(x = min_ + timedelta(minutes = 5*1440), c = 'k', ls = (0, (20, 6)), lw = 0.5)
    plt.axvline(x = min_ + timedelta(minutes = 6*1440), c = 'k', ls = (0, (20, 6)), lw = 0.5)
    # plt.axvline(x = min_ + timedelta(minutes = 7*1440), c = 'k', ls = '--', lw = 0.5)
    
    plt.text(min_ + timedelta(minutes = 1440/2), 0.7, 'Day 1', fontsize = 9, ha = 'center', va = 'center')
    plt.text(min_ + timedelta(minutes = 3*1440/2), 0.7, 'Day 2', fontsize = 9, ha = 'center', va = 'center')
    plt.text(min_ + timedelta(minutes = 5*1440/2), 0.7, 'Day 3', fontsize = 9, ha = 'center', va = 'center')
    plt.text(min_ + timedelta(minutes = 7*1440/2), 0.7, 'Day 4', fontsize = 9, ha = 'center', va = 'center')
    plt.text(min_ + timedelta(minutes = 9*1440/2), 0.7, 'Day 5', fontsize = 9, ha = 'center', va = 'center')
    plt.text(min_ + timedelta(minutes = 11*1440/2), 0.7, 'Day 6', fontsize = 9, ha = 'center', va = 'center')
    plt.text(min_ + timedelta(minutes = 13*1440/2), 0.7, 'Day 7', fontsize = 9, ha = 'center', va = 'center')
    
    plt.legend(bbox_to_anchor=(0.75, 1.16), ncol = 2, fontsize = 12, frameon = True, edgecolor = 'black', framealpha=1)
    
    plt.savefig(r'C:/PhD Research/Airborne/Processed/plots/Figure_s5_%s.jpg' %v, format='jpg', dpi=1600, bbox_inches='tight')
    
```
